```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
```

<h3> vidaccs </h3>

```python
seeds = [2,3,6]

vid_accs = []
total_accs = []
for seed in seeds:
    vid_accs.append(np.loadtxt(f'/tmp/seed_{seed}.txt'))
    total_accs.append(np.loadtxt(f'/tmp/seed_avg_{seed}.txt'))

mean_acc = np.mean(total_accs)
mean_acc_per_vid = np.mean(vid_accs, axis=0)

print(total_accs)
total_std = np.std(total_accs)
per_vid_std = np.std(vid_accs, axis=0)

print(mean_acc)
print(mean_acc_per_vid)
print('-------')
print(total_std*100)
print(per_vid_std*100)

plt.title('R3D')
plt.ylim(0,1)
plt.xticks(range(0, len(mean_acc_per_vid)+1), [str(i) for i in range(1, len(mean_acc_per_vid)+1)] + ['Avg.'])
bars1 = plt.bar(range(len(mean_acc_per_vid)), mean_acc_per_vid)
bars2 = plt.bar(7, mean_acc)

for i, bar in enumerate(bars1):
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .006, f'{yval*100:02.1f}%')

for i, bar in enumerate(bars2):
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .006, f'{yval*100:.1f}%')
```
